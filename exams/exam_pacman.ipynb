The dimensionality of this pacman numpy.array represent the width, height and the number of color channel (red, green, blue).

Fun fact, it you find a RGB-A image in the format of (200,200,4), then the 4th channel of the image represent the alpha value.(For those interested, plz google what the alpha value mean in the context of image processing)

```python
## use plt.imshow() to show see the 
plt.imshow(pacman)
```

## now if we only visualize the red, green and blue channal separately [R,G,B]

```python
plt.subplot(131)
plt.imshow(pacman[:,:,0])
plt.title('red channel')
plt.subplot(132)
plt.imshow(pacman[:,:,1])
plt.title('green channel')
plt.subplot(133)
plt.imshow(pacman[:,:,2])
plt.title('blue channel')
```

We see that the blue channel doesn't contribute to the image at all
